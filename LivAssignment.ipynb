Que3 Explain the following terms

 The difference in np.array, np.asarray and np.asanyarray
 
 The difference between Deep copy and shallow copy

np.asarray converts the input to an array, avoiding copies whenever possible.

np.array creates a new array object. It can optionally copy the data if the input is already an array but has a different data type or if copy=True is explicitly specified.

np.asanyarray converts input to an array but allows for the creation of arrays with subclasses (e.g., matrix or masked arrays).
    

A shallow copy creates a new object, but the contents (elements) inside the object are references to the original elements.

A deep copy creates a new object and recursively copies all objects inside it.

```python


# Create a (5,6) array with random integers between 1 and 10
array = np.random.randint(1, 11, (5, 6))
print("Original Array:\n", array)

# a) Extract all even integers from the array
even_integers = array[array % 2 == 0]
print("Even Integers:\n", even_integers)

# b) Extract all odd integers from the array
odd_integers = array[array % 2 != 0]
print("Odd Integers:\n", odd_integers)
```

Que14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and 
Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and 
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
dates = pd.date_range(start='2024-01-01', end='2024-01-10', freq='D')

# Generate random data for 'Temperature' and 'Humidity'
temperature = np.random.uniform(15, 30, size=len(dates))  # Temperature between 15 and 30 degrees
humidity = np.random.uniform(40, 80, size=len(dates))     # Humidity between 40% and 80%

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```

Que15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following 
tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Task: Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# Task (a): Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Task (b): Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim()  # Get the range of x-axis
x = np.linspace(xmin, xmax, 100)  # Create x-values for the PDF line plot
pdf = norm.pdf(x, loc=0, scale=1)  # Calculate the PDF values for a normal distribution
plt.plot(x, pdf, 'k', linewidth=2, label='Normal Distribution PDF')

# Task (c): Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Task (d): Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

Que17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the 
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python


# Create two random arrays of equal length
x = np.random.randn(100)  # Random data for x-axis
y = np.random.randn(100)  # Random data for y-axis

# Determine the quadrant of each point
# Quadrant 1: x > 0 and y > 0
# Quadrant 2: x < 0 and y > 0
# Quadrant 3: x < 0 and y < 0
# Quadrant 4: x > 0 and y < 0
quadrants = []

for xi, yi in zip(x, y):
    if xi > 0 and yi > 0:
        quadrants.append('Quadrant 1')
    elif xi < 0 and yi > 0:
        quadrants.append('Quadrant 2')
    elif xi < 0 and yi < 0:
        quadrants.append('Quadrant 3')
    else:
        quadrants.append('Quadrant 4')

# Create a DataFrame for Seaborn
data = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Create a Seaborn scatter plot, color points based on 'Quadrant'
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='coolwarm')

# Add labels for the axes and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrants')

# Show the plot
plt.show()
```

Que18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine 
Wave Function'8

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)  # 100 points between 0 and 4π
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='sin(X)', tools="pan,box_zoom,reset")

# Plot the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Add grid lines
p.grid.grid_line_alpha = 0.3  # Make grid lines semi-transparent

# Show the plot
show(p)
```

Que19. generate a bar chart of randomly generated categorical data, color bars based on their 
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical 
Bar Chart'.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)  # Random values between 10 and 100 for the categories

# Create a DataFrame (optional but useful for organizing data)
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create ColumnDataSource for Bokeh
source = ColumnDataSource(data)

# Create a figure object
p = figure(x_range=categories, title="Random Categorical Bar Chart", 
           toolbar_location=None, tools="hover", tooltips="@Category: @Value", 
           x_axis_label='Categories', y_axis_label='Values')

# Create the bar chart
p.vbar(x='Category', top='Value', width=0.5, source=source, color='blue', legend_field="Category")

# Customize the bar color based on values (darker color for higher values)
p.vbar(x='Category', top='Value', width=0.5, source=source, 
       color='color', legend_field="Category")

# Add hover tooltips to show values
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Set grid and labels
p.grid.grid_line_alpha = 0.3

# Show the plot
output_file("categorical_bar_chart.html")
show(p)
```
